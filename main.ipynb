```python
import numpy as np
from scipy.io import wavfile

cons = [[0.7217, 1.0247],
        [0.5346, 0.9273],
        [0.5346, 1.0247],
        [0.5346, 1.1328],
        [0.5906, 0.9273],
        [0.5906, 1.0247],
        [0.5906, 1.1328],
        [0.6535, 0.9273],
        [0.6535, 1.0247],
        [0.6535, 1.1328]]

sample_rate = 8192
n = np.arange(0, 1000)

NUM = 0
for w in cons:
    data = np.sin(w[0] * n) + np.sin(w[1] * n)
    
    data /= np.max(np.abs(data))
    data = (data * 32767).astype(np.int16)

    wavfile.write(f"{NUM}.wav", sample_rate, data)
    NUM += 1
```

```python
def get_dominant_freqs(signal, fs=8192, top_k=2):
    N = 2048
    padded = np.pad(signal, (0, N - len(signal)), mode='constant')
    spectrum = np.abs(fft(padded))[:N//2]
    freqs = np.fft.fftfreq(N, 1/fs)[:N//2]

    top_indices = np.argsort(spectrum)[-top_k:]
    top_freqs = sorted([round(freqs[i]) for i in top_indices])
    return top_freqs

db = {}
for i in range(10):
    sr, ref_sig = wavfile.read(f"{i}.wav")
    ref_sig = ref_sig.astype(np.float32)
    ref_sig /= np.max(np.abs(ref_sig))
    freqs = get_dominant_freqs(ref_sig, fs=sr)
    db[i] = freqs
```

```python
def match_digit(freqs, ref_db):
    def score(ref):
        return sum(abs(a - b) for a, b in zip(freqs, ref))
    return min(ref_db.items(), key=lambda item: score(item[1]))[0]

digits = []
for i in range(7):
    start = i * (1000 + 100)
    end = start + 1000
    segment = x1[start:end]
    segment = segment.astype(np.float32)
    segment /= np.max(np.abs(segment))
    freqs = get_dominant_freqs(segment, sample_rate)
    digit = match_digit(freqs, db)
    digits.append(str(digit))

num1 = ''.join(digits)
print(num1)
```
